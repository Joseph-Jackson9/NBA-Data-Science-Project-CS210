```python
import pandas as pd
import requests
pd.set_option('display.max_columns',None)
import time
import numpy as np
```

```python

# this is the main URL for the page we will be scraping. we will be looping through each page and will be changing key aspects of the URL later.
test_url ='https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season=2023-24&SeasonType=Regular%20Season&StatCategory=PTS'
```

```python
r = requests.get(url=test_url).json()
headers = r['resultSet']['headers'] 
pd.DataFrame(r['resultSet']['rowSet'],columns = headers)
```

```python
#getting the header for the table set up
df_cols = ['Year','Season_type']+ headers
pd.DataFrame(columns=df_cols)
```

```python
df = pd.DataFrame(columns=df_cols)
season_types = ['Regular%20Season','Playoffs']
years = ['2013-14','2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23', '2023-24']
begin_loop= time.time()

for y in years:
    for s in season_types:
        api_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season='+y+'&SeasonType='+s+'&StatCategory=PTS'
        r= requests.get(url=api_url).json()
        temp_df1 = pd.DataFrame(r['resultSet']['rowSet'],columns = headers)
        temp_df2 = pd.DataFrame({'Year':[y for i in range(len(temp_df1))],
                                'Season_type':[s for i in range(len(temp_df1))]})
        temp_df3 = pd.concat([temp_df2,temp_df1], axis = 1)
        df = pd.concat([df,temp_df3], axis = 0)
        print(f'Finished scraping data for the {y} {s}')
        lag = np.random.uniform(low = 4, high = 30)
        print(f'... waiting {round(lag,1)} seconds.')
        time.sleep(lag)

print(f'process completed. Total runtime: {round((time.time() - begin_loop)/60 ,2)} minutes.')
df.to_excel('Tools/nba_player_data2.xlsx',index = False)
```
